```python
import pandas as pd
df = pd.DataFrame(covid.get_data())
print(df)
```

```python
sorted = df.sort_values(by=['confirmed'], ascending=False)
```

```python
excluded = sorted[~sorted.country.isin(['Europe', 'South', 'America', 'Asia', 'World', 'North America', 'Africa'])]
```

```python
top10 = excluded.head(10)
print(top10)
```

```python
x = top10.country
y1 = top10.confirmed
y2 = top10.active
y3 = top10.deaths
y4 = top10.recovered
```

```python
%matplotlib inline
import matplotlib.pyplot as pl
```

```python
pl.plot(x, y1) #Create a simple line graph
pl.xticks(rotation=90)
pl.show() 
```

```python
pl.bar(x, y1) #We can display the bar graph
pl.xticks(rotation=90)
pl.show()
```

```python
pl.plot(x, y1, label='Confirmed') #Use this code to create a multiline graph
pl.plot(x, y2, label='Active')
pl.plot(x, y3, label='Deaths')
pl.plot(x, y4, label='Recovered')
pl.legend(loc='upper right')
pl.xticks(rotation=90)
pl.show()
```

```python
labels = ['Confirmed', 'Active', 'Deaths', 'Recovered'] #We can rewrite the preceding example with compact code and better formatting
pl.plot(x, y1, x, y2, x, y3, x, y4)
pl.legend(labels, loc='upper right')
pl.xticks(rotation=90)
pl.grid()
pl.show()
```

```python
df2 = pd.DataFrame([y1, y2, y3, y4]) #We can display the data in vertical multiple bar graphs
df2.plot.bar();
pl.legend(x, loc='upper center')
pl.xticks(rotation=90)
pl.grid()
pl.show()
```

```python
df2.plot.bar(stacked=True);  #We can plot the data in vertical stacked bar graphs
pl.legend(x, loc='upper center')
pl.xticks(rotation=90)
pl.grid()
pl.show()
```

```python
df2.plot.barh();  #We can create horizontal bar graphs
pl.legend(x, loc='upper right')
pl.xticks(rotation=90)
pl.grid()
pl.show()
```

```python
df2.plot.barh(stacked=True); #create a stacked horizontal bar visualization
pl.legend(x, loc='upper right')
pl.xticks(rotation=90)
pl.grid()
pl.show()
```

```python
df2.plot.area();  #We can even use area graphs to visualize this data. By default, the area graph, as we know, is stacked.
pl.legend(x, loc='upper right')
pl.xticks(rotation=90)
pl.grid()
pl.show()
```

```python
df2.plot.area(stacked=False); #create an overlapping area graph
pl.legend(x, loc='upper right')
pl.xticks(rotation=90)
pl.grid()
pl.show()
```

```python
factor=0.00001  #create a nice scatter plot where the size of dots is proportional to the magnitude of data
pl.scatter(x, y1, s=y1*factor);
pl.scatter(x, y2, s=y2*factor);
pl.scatter(x, y3, s=y3*factor);
pl.scatter(x, y4, s=y4*factor);
pl.legend(labels, loc='upper right')
pl.xticks(rotation=90)
pl.grid()
pl.show()
```

```python
pl.pie(y1, labels=x)  #create a nice pie chart
pl.title('Confirmed Cases')
pl.show()
```
